```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# Scatter Plot

```python
x=np.arange(0,20,2)
y=np.arange(20,30,1)
```

```python
plt.scatter(x,y)
```

# plot

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'g*')
```

```python
plt.plot(x,y,'bo')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("2d graph")
```

# subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'b+')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

# sine wave

```python
np.pi
```

```python
x=np.arange(0, 4 * np.pi, 0.1)
y=np.sin(x)
plt.title('sine wave')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,10)
y=np.sin(x)
plt.title('sine wave')
plt.plot(x,y)
plt.show()
```

# Bar

```python
x=np.arange(0,10)
y=np.arange(10,20)
x1=np.arange(20,30)
y2=np.arange(30,40)
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
x1=np.arange(10,20)
y2=np.arange(10,20)
plt.title("bar graph")
plt.bar(x,y,align="center")
plt.bar(x1,y2,color='g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Histogram

```python
a=np.arange(1,20)
b=np.arange(20,30)
```

```python
plt.hist(a,bins=5)
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title("histogram")
```

# BoxPlot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart

```python
labels="Python","C++","Java","PHp"
explode=(0.4,0.1,0,0)
colors=["gold","yellowgreen","lightcoral","lightskyblue"]
sizes=[215,312,512,612]
```

```python
plt.pie(sizes,explode=explode, labels=labels,colors=colors,autopct="1%1.1f%%",shadow=True)
```
